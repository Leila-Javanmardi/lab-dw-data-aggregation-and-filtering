2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
da_yes= da_ma[(da_ma.response=='Yes')]
da_yes.head(4)
da_yes_filter= da_yes.pivot_table(index='gender', columns='policy_type', values='total_claim_amount',aggfunc='mean').round(2)
display(da_yes_filter)
```

2- Overall, males tend to have a slightly lower average total claim amount except for the "Personal Auto".
